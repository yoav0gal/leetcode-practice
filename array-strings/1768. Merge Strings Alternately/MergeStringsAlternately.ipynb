### My plan:
1. Iterate over both strings until both of them got pass their last char.
2. If the string still have a char add the char, else add a space.
3. Return the result.

```python
def mergeAlternately( word1: str, word2: str) -> str:
    i = 0
    word1Char= word1[i]
    word2Char= word2[i]
    mergedWord =""


    while((word2Char != "" or word1Char != "") ): 
        mergedWord += (word1Char+word2Char)
        i = i+1
        word1Char = word1[i] if word1.__len__() > i else ""
        word2Char = word2[i] if word2.__len__() > i else ""

    return mergedWord
```

```python
print(mergeAlternately("abc", "pqr"))
print(mergeAlternately("ab", "qprs"))
print(mergeAlternately("abcd", "pq"))
```
